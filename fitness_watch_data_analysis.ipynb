## 2. Data Cleaning and Organization üßπ

- **Convert Date Column:**  
  Transform the 'Date' column into datetime format for easier time-based analysis.

- **Check for Missing Values:**  
  Identify any null or missing entries in the dataset to ensure data quality.

These steps ensure the dataset is clean, consistent, and ready for meaningful analysis.


````python
import pandas as pd

# Step 1: Load data
df = pd.read_csv(r"D:\Fitness Watch Data Analysis\Fitness_Watch_Data_Analysis_200.csv")
print(df.head())  # Shows first 5 rows

# Step 2: Clean / Organise
df['Date'] = pd.to_datetime(df['Date'])

missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)
````

## 3. Exploratory Data Analysis (EDA) üîç

- **Summary Statistics:**  
  Generate descriptive statistics to understand data distribution and central tendencies.

- **Unique Users:**  
  Count the number of unique users in the dataset for user diversity insight.

- **Correlation Analysis:**  
  Calculate correlation matrix to find relationships between numerical features.


````python
# Step 3: EDA

# Summary statistics
summary = df.describe()
print("Summary Statistics:\n", summary)

# Count unique users
unique_users = df['User_ID'].nunique()
print("\nNumber of Unique Users:", unique_users)

# Correlation matrix
correlation = df.corr()
print("\nCorrelation Matrix:\n", correlation)
````

## 4. Visual Representation üìä

- **Scatter Plot (Steps vs Calories Burned):**  
  Visualize how step counts affect calories burned.

- **Histogram (Sleep Hours Distribution):**  
  Understand the distribution and common sleep durations among users.

- **Heatmap (Feature Correlation):**  
  Display correlations among features for better insight into variable interactions.


````python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot: Steps vs Calories Burned
sns.scatterplot(data=df, x='Steps', y='Calories_Burned')
plt.title("Steps vs Calories Burned")
plt.show()

# Histogram: Sleep Hours Distribution
sns.histplot(df['Sleep_Hours'], kde=True)
plt.title("Sleep Hours Distribution")
plt.show()

# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
````
